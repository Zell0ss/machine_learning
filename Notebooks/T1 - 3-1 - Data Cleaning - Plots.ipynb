## Histograma de frecuencia
La *regla de Sturges*, propuesta por Herbert Sturges en 1926, es una regla pr√°ctica acerca del n√∫mero de clases que deben considerar al elaborarse un histograma.

Este n√∫mero viene dado por la siguiente expresi√≥n:
````
c=1+(log(M)/log(2))
````
o en log2
````
c=1+log2(M)
````
donde M es el tama√±o de la muestra. Si no ponemos nada en bins, hist ser√° pintado con la regla de sturges

`````python
sturges = 1 + np.log2(data["Day Calls"].count())

plt.hist(data["Day Calls"], bins = int(sturges))
plt.xlabel("Num. Llamadas/dia")
plt.ylabel("Frecuancia en los datos")
plt.title("Histograma de n√∫mero de llamadas al d√≠a")
plt.savefig(dspath+"/customer-churn-model/4.jpeg")
`````
