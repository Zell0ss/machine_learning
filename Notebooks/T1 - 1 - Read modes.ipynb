## Carga de datos con fn open


```python
data4 = open(os.path.join(dspath,"customer-churn-model/Customer Churn Model.txt"), "r" )
cols = data4.readline().strip().split(",")
n_cols = len(cols)

#relleno diccionario vacio
counter = 0 
main_dict = {}
for col in cols:
    main_dict[col] = []

for line in data4:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter +=1

print ("El dataset tiene {} filas y {} columnas".format(counter, n_cols))
    
```

```python
df4 = pd.DataFrame(main_dict)
df4.head()
```

## Creacion de fichero de tabs

```python
infile = os.path.join(dspath,"customer-churn-model/Customer Churn Model.txt")
outfile = os.path.join(dspath,"customer-churn-model/Tabb Customer Churn Model.txt")

with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
            
df5 = pd.read_csv(outfile, sep = "\t")
df5.head()
```
