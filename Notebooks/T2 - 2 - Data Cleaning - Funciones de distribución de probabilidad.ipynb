# Funciones de distribución de probabilidades

## Distribución uniforme

```python
import numpy as np
import matplotlib.pyplot as pyplt
```

```python
linf = 1
lsup = 100
nmuestras = 1000

data = np.random.uniform(linf, lsup, nmuestras)
```

```python
pyplt.plot(x,sorted(data))
```

```python
pyplt.hist(data)
```

## Distribución normal o de gauss

```python
data = np.random.randn(1000)
```

```python
x = range(1, 1001)
pyplt.plot(x, data)
```

```python
pyplt.hist(data)
```

```python
pyplt.plot(x,sorted(data))
```

```python
mu = 5.5
sd = 2.5
data = mu + sd*np.random.randn(1000) # Z = (X-mu) / sd
```

```python
pyplt.hist(data)
```
