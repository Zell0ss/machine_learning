## La simulación de Monte Carlo

Vamos a usar la simulación de montecarlo para calcular pi.
Dado un círculo de radio r inscrito en un cuadrado de lado 2r la probailidad de caer en el círculo es:

```` python
P(caer dentro cisrculo) = caer dentro círculo / caer dentro del cuadrado = Pi*r*r/2*r*2*r = Pi/4
````
El círculo en cuestión cumple 
```` x*x + y*y = r*r````
El cuadrado cumple 
````x = 2r, y = 2r````

Si centramos el circulo en 0,0 y de radio 1 y cogemos sólo x e y entre 0 y 1 

* Generamos una lista de 1100 parejas de num aleatorios uniformes entre 0 y 1
 
* Por cada pareja calcularemos x*x + y*y
     - si sale menor de 1 (1*1): dentro del primer cuadrante del círculo
     - si sale mayor de 1: fuera del primer cuadrante del círculo
     
* Calculando el número total de veces que están dentro del círculo y lo dividimos enter el total para obtener una aproximación de la P(caer dentro)

* Usamos dicha probabilidad para aproximar Pi

* Repetimos un numero suf. de veces (1000) para otener (1000) diferentes aproximaciones de Pi

* Calculamos Pi como el promedio de estos valores


`````python
pi_avg = 0 
parejas = 10000
pi_value_list = []
experimentos = 1000
for i in range(experimentos):
    dentro = 0
    x = np.random.uniform(0,1,parejas).tolist()
    y = np.random.uniform(0,1,parejas).tolist()
    for j in range(parejas):
        z=x[j]*x[j]+y[j]*y[j]
        if (z<=1):
            dentro += 1
    float_dentro = float(dentro)
    pi_value = float_dentro *4 / parejas
    pi_value_list.append(pi_value)
    pi_avg += pi_value

pi = pi_avg/experimentos

print (pi)
pyplt.plot(pi_value_list)
`````
