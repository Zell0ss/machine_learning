## A concatenar.

```python
import pandas as pd
import os
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])



df2 = pd.DataFrame({'E': ['E3', 'E5', 'E6', 'E7'],
                    'F': ['F3', 'F5', 'F6', 'F7'],
                    'G': ['G3', 'G5', 'G6', 'G7']},
                    index=[3, 5, 6, 7])
```

```python
pd.concat([df1, df2], axis=1)
```

### reseteando Ã­ndices

```python
def concatenawer (sb1, sb2):
    mydf1= sb1.copy()
    mydf2= sb2.copy()
    mydf1.reset_index(drop=True, inplace=True)
    mydf2.reset_index(drop=True, inplace=True)
    return pd.concat( [mydf1, mydf2], axis=1)

concatenawer(df1, df2)
```
