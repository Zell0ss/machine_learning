## Datos distribuidos
En eldirectorio distributed data tenemos +300 ficheros con cuatro columnas con la calidad del aire. Los datos ademas vienen muy distribuidos asi que la idea es concatenarlos para manejar un único dataset.

```python
aire1 = pd.read_csv("../../datasets/distributed-data/001.csv")
aire1.head()
```

```python
aire1.describe()
```

Fijandonos en el count vemos que de los 1461 filas sólo117 tienen datos de sulfate y solo 112 de nitrate: es una matriz sparse con datos muy espaciados entre ellos.
La idea es usar aire1 como base y concatenar detras los ficheros uno a uno

```python
filepath = "../../datasets/distributed-data/"
aire = pd.read_csv("../../datasets/distributed-data/001.csv")
final_lenght = len (aire)

for i in range (2, 333):
    s = "000{}".format(i)
    s =  filepath + s[-3:] + ".csv"
    temp = pd.read_csv(s)
    final_lenght += len(temp)
    aire = pd.concat([aire, temp], axis = 0)
print (final_lenght)
aire1.describe()
```

## Joins de datasets
Basicamente es igual a los join de bases de datos. 

### inner join

```python
filepath = "../../datasets/athletes/"
data_main = pd.read_csv(filepath+"Medals.csv", encoding = "ISO-8859-1" )
data_main.head()
```

```python
data_main["Athlete"].unique()
```

```python
data_country = pd.read_csv(filepath+"Athelete_Country_Map.csv",encoding = "ISO-8859-1"  )
data_country.describe()
```

```python
data_sport = pd.read_csv(filepath+"Athelete_Sports_Map.csv",encoding = "ISO-8859-1"  )
data_sport.describe()
```

### joins con pd.merge

```python
data_country = data_country.drop_duplicates(subset="Athlete")
```

```python
data_main_country = pd.merge(left = data_main, right = data_country, left_on="Athlete", right_on="Athlete" )

data_main_country.head()
```

```python
data_sport = data_sport.drop_duplicates(subset="Athlete")
```

```python
data_all = pd.merge(left = data_main_country, right = data_sport, left_on="Athlete", right_on="Athlete" )
data_all.head(25)
```
