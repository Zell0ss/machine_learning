```python
import nltk
from bs4 import BeautifulSoup
from urllib import request
nltk.download()
```

```python
response = request.urlopen('https://es.wikipedia.org/wiki/Infierno_(Divina_comedia)')
html = response.read()
print (html)
```

## To clean
Para limpiar los tags html y demás basura tenemos BeautifulSoup. Ojo, si no cargaste las librerias previamente, cargalas ahora y **reinicia jupyter**

```python

soup = BeautifulSoup(html,"html5lib")
# soup = BeautifulSoup(html,"lxml")
# soup = BeautifulSoup(html,"html.parser")
text = soup.get_text(strip=True)

print (text)
```

## tokenizar con la función split
Es lo más básico, pero también lo más bárbaro: separamos por espacios y a correr.

```python

tokens = [t for t in text.split()]
print (tokens)
```

## distribución de frecuencia

```python
freq = nltk.FreqDist(tokens)

for key,val in freq.items():
    print (str(key) + ':' + str(val))
```

```python
freq.plot(40, cumulative=False)
```

## Eliminar palabras *de parada* con NLTK

NLTK contiene listas de palabras prohibidas para la mayoría de los idiomas. Para obtener las palabras de parada en español:

```python
from nltk.corpus import stopwords

sr = stopwords.words('spanish')

## pinto las diez primeraws como ejemplo
sr[:10]
```

Modificando para limpiarlas: debemos comparar los tokens en minusculas

```python
clean_tokens = tokens[:]
for token in tokens:
    if token.lower() in sr:
        clean_tokens.remove(token)
freq_clean = nltk.FreqDist(clean_tokens)
freq_clean.plot(40, cumulative=False)
```

## Tokenizar usando NLTK

Vimos cómo dividir el texto en tokens utilizando la función split; ahora veremos cómo convertir el texto en tokens con NLTK.

Tokenizar el texto es importante ya que el texto no se puede procesar sin tokenizarlo. El proceso de tokenización significa dividir partes más grandes en partes pequeñas.

Puedes convertir párrafos en oraciones y confeccionar frases en palabras según tus necesidades. NLTK contiene un tokenizador de frase y un tokenizador de palabra.

Supongamos que tenemos un texto de muestra como este mismo parrafo (muy meta)

Para frases tenemos sent_tokenizer, para palabras word_tokenizer. Saben que Sr. Sra. Mr. etc. son palabras aunque contengan un punto, y no indicanm fin de frase

```python


mitexto = "## Tokenizar usando NLTK \
Vimos cómo dividir el texto en tokens utilizando la función split; ahora veremos cómo convertir el texto en tokens con NLTK. \
Tokenizar el texto es importante ya que el texto no se puede procesar sin tokenizarlo. El proceso de tokenización significa dividir partes más grandes en partes pequeñas. \
Puedes convertir párrafos en oraciones y confeccionar frases en palabras según tus necesidades. El Sr. NLTK contiene un tokenizador de frase y un tokenizador de palabra. \
Supongamos que tenemos un texto de muestra como este mismo parrafo (muy meta)"

```

```python
from nltk.tokenize import sent_tokenize

print(sent_tokenize(mitexto))
```

```python
from nltk.tokenize import word_tokenize

print(word_tokenize(mitexto))
```

## Obtener sinónimos (un synset) de WordNet

Si recuerdas, instalamos paquetes NLTK usando nltk.download(). Uno de los paquetes fue WordNet.
https://www.nltk.org/howto/wordnet.html

WordNet es una base de datos creada para el procesamiento del lenguaje natural. Incluye grupos de sinónimos y una breve definición.

Puedes obtener estas definiciones y ejemplos para una palabra dada como esta:
https://likegeeks.com/es/tutorial-de-nlp-con-python-nltk/

```python
from nltk.corpus import wordnet as wn
print([snset.lemma_names('spa') for snset in wn.synsets('placer', lang='spa')])
print(wn.synsets('placer', lang='spa')[0].definition())
print(wn.synsets('placer', lang='spa')[0].examples())
```
